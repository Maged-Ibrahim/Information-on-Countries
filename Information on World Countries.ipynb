```python
data = pd.read_csv('D:Maged/countries/countries of the world.csv', decimal=',')
print('Number of missing data "MD" :')
print(data.isnull().sum())
data.describe(include='all')
# Number of missing data in the dataset.
```

```python
data.groupby('Region')[['GDP ($ per capita)', 'Literacy (%)', 'Agriculture']].median()
#  Filling the missing data using the median of the region that a country belongs "From Kaggle Notebook"
```

```python
for col in data.columns.values:
    if data[col].isnull().sum() == 0:
        continue
    if col == 'Climate':
        guess_values = data.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        guess_values = data.groupby('Region')[col].median()
    for region in data['Region'].unique():
        data[col].loc[(data[col].isnull())&(data['Region']==region)] = guess_values[region]
```

```python
print(data.isnull().sum())
# All missing values are filled successfully.
```

```python
data.loc[(data['Deathrate']>20) & (data['GDP ($ per capita)'] < 15000)]
# The relationship between High "Deathrate" and "GDP ($ per capita)".
```

```python
data.loc[(data['Climate']<10) & (data['GDP ($ per capita)'] < 15000)]
# The relationship between "Climate" and "GDP ($ per capita)". Climate could affect the crops and so the GDP.
```

```python
data.loc[(data['Crops (%)']>10) & (data['GDP ($ per capita)'] < 15000)]
# The relationship between High "Crops" and "GDP ($ per capita)". More Crops = more food. more food reserve could affect the GDP.
```
